# K Means Clustering

## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing dataset

```python
data_set = pd.read_csv("Amazon.com Model.csv")
X = data_set.iloc[:, [2,4]].values
```

## Optimal number of clusters via Elbow Method

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=21)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title("WCSS via Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

## K Means Model Training on Training set

```python
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=42)
y_means = kmeans.fit_predict(X)
```

## Visualizing Clusters

```python
plt.scatter(X[y_means == 0, 0] , X[y_means == 0, 1], s=100, c='violet', label='Cluster1')
plt.scatter(X[y_means == 1, 0] , X[y_means == 1, 1], s=100, c='pink', label='Cluster2')
plt.scatter(X[y_means == 2, 0] , X[y_means == 2, 1], s=100, c='blue', label='Cluster3')
plt.scatter(X[y_means == 3, 0] , X[y_means == 3, 1], s=100, c='cyan', label='Cluster4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='orange', label='Centroids')
plt.title("Clusters of Amazon Users based on Age")
plt.xlabel("Age")
plt.ylabel("Purchase Ratings")
plt.legend()

plt.show()
```
