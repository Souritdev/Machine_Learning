# Hierarchical Clustering

## Importing libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

## Importing dataset

```python
data_set = pd.read_csv("Amazon.com User Clustering.csv")
X = data_set.iloc[:, [2,4]].values
```

## Optimal number of clusters via DendroGram

```python
import scipy.cluster.hierarchy as sch 
dendrograms = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrograms")
plt.xlabel("Age")
plt.ylabel("Euclidean Distance")
```

## Hierarchical Clustering Model Training on Training set

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
yhc = hc.fit_predict(X)
```

## Visualizing Clusters

```python
plt.scatter(X[yhc == 0, 0] , X[yhc == 0, 1], s=100, c='magenta', label='Cluster1')
plt.scatter(X[yhc == 1, 0] , X[yhc == 1, 1], s=100, c='red', label='Cluster2')
plt.scatter(X[yhc == 2, 0] , X[yhc == 2, 1], s=100, c='blue', label='Cluster3')
plt.scatter(X[yhc == 3, 0] , X[yhc == 3, 1], s=100, c='cyan', label='Cluster4')
plt.scatter(X[yhc == 4, 0] , X[yhc == 4, 1], s=100, c='cyan', label='Cluster5')


plt.title("Amazon Users")
plt.xlabel("Age")
plt.ylabel("Purchase Ratings")
plt.legend()

plt.show()
```
