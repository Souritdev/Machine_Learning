# Random Forest Regression

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing Data Set

```python
data_set = pd.read_csv("Salary_DataSetsReg.csv")
X = data_set.iloc[:, 1:-1].values
y = data_set.iloc[:, -1].values
```

## Training the dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=12, random_state=42)
regressor.fit(X,y)
```

## Predicting a new result

```python
regressor.predict([[5.5]])
```

## Visualising the Random Forest Regression

```python
X_grid = np.arange(min(X), max(X) , 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y , color = 'Red')
plt.plot(X_grid, regressor.predict(X_grid), 1)
plt.title("Random Forest Regression")
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()
```
